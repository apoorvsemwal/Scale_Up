```python
#Handling Multiple Data Frames
#right_df = pd.DataFrame({'key1':['a','a','b','b'], 'key2':['one','one','one','two'], 'rval'})
```

```python
df_join1 = pd.DataFrame({'key':['x','y','z'], 'lval':[1,2,9]})
df_join2 = pd.DataFrame({'key':['x','y','w'], 'rval':[3,4,99]})
```

```python
pd.merge(df_join1,df_join2)# Take Common elements based on coloumnn names here only 'key' coloums is same so its on the basis of that
```

```python
pd.merge(df_join1,df_join2, how="left")# Take Left Inner Join
```

```python
pd.merge(df_join1,df_join2, how="right")# Take Right Inner Join
```

```python
pd.merge(df_join1,df_join2, how="outer")# Take Outer
```

```python
df_join3 = pd.DataFrame({'key1':['x','x','y','z'], 'lval':[1,3,2,9]})
df_join4 = pd.DataFrame({'key2':['x','x','y','w'], 'rval':[3,4,44,99]})
```

```python
pd.merge(df_join3,df_join4, how="left", left_on="key1", right_on="key2") #Now we are telling the field names
```
